### 전체 댓글

```python
code_b, ep_num_b, comments_b = [], [], []
```

```python
# 웹툰 detail - 코드, 회차, 베스트 댓글

driver = webdriver.Chrome('./chromedriver')

for i in df_list['code'][202:]:
    start = time.time()
    
    for j in range(int(df_list[df_list['code'] == i]['max_pg'])):
        url = 'https://comic.naver.com/webtoon/detail?titleId=' + str(i) + '&no=' + str(j + 1)
        driver.get(url)
        driver.switch_to.frame('commentIframe')
        
        for k in range(15):
            code_b.append(i)
            ep_num_b.append(j + 1)
            comments_b.append(driver.find_elements_by_css_selector('.u_cbox_text_wrap')[k].text)
                
    print(i, f"{time.time() - start:.0f} s")
```
