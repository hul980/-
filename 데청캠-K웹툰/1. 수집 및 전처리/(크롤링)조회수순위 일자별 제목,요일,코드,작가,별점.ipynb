```python
#오늘 날짜
import datetime as dt
now=dt.datetime.now()
today = now.strftime('%Y-%m-%d')
today

import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
title=[]
code=[]
author=[]
rate=[]
week=[]

#요일별웹툰
for i in ['mon', 'tue', 'wed', 'thu', 'fri', 'sat', 'sun']:
    raw = requests.get('https://comic.naver.com/webtoon/weekdayList?week='+i+'&order=ViewCount&view=image')
    html = BeautifulSoup(raw.text,'html.parser')
    
    #제목
    raw_title_list = html.select('dt')
    for j in raw_title_list[3:] : 
        title.append(j.get_text())
    
    container = html.select('ul.img_list li')
    for cont in container:
        #작가
        author.append(cont.select_one('dd.desc a').text)
        #평점
        rate.append(cont.select_one('div.rating_type strong').text)
        #코드
        code.append(re.sub(r'[^0-9]', '', str(cont.select_one('dt'))[:45]))
        #요일
        week.append(i)

list_df = pd.DataFrame({'제목' : title,
                        '요일': week,
                       '코드': code,
                       '작가': author,
                       '별점': rate})
list_df
```

```python
list_df.to_csv(today+'조회순기준 제목,요일,코드,작가,별점.csv', index=False, encoding='cp949')
```
