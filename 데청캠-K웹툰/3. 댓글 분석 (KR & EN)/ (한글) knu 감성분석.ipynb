```python
import re, time, requests, json
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

with open('/content/drive/MyDrive/Colab Notebooks/SentiWord_info.json', encoding='utf-8-sig', mode='r') as f:
    data = json.load(f)

dict_knu = pd.DataFrame(data)
dict_knu
```

```python
start = time.time()

kw_2_comment = pd.read_csv('kw_2_comment.csv', encoding='UTF-8')

score_list = []

#score_sum, score_two, score_one, score_zero, score_mone, score_mtwo = [], [], [], [], [], []

for i in range(len(kw_2_comment['best_comment'][:1])):
    
    score_sum, score_two, score_one, score_zero, score_mone, score_mtwo = 0, 0, 0, 0, 0, 0
    
    for j in range(len(kw_2_comment['best_comment'][i].split(' '))):
    
        for k in range(0, len(dict_knu)):

            if dict_knu['word'][k] == kw_2_comment['best_comment'][i].split(' ')[j]:
                
                score_sum += int(dict_knu['polarity'][k])
                
                if int(dict_knu['polarity'][k]) == 2:
                    score_two += 1
                    
                elif int(dict_knu['polarity'][k]) == 1:
                    score_one += 1
                    
                elif int(dict_knu['polarity'][k]) == 0:
                    score_zero += 1
                    
                elif int(dict_knu['polarity'][k]) == -1:
                    score_mone += 1
                    
                else:
                    score_mtwo += 1
                
    score_list.append([score_sum, score_two, score_one, score_zero, score_mone, score_mtwo])

print(f"{time.time() - start:.0f} s")
```
